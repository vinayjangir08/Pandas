```python
import pandas as pd
```

```python
df = pd.DataFrame({'id':[101,102,103,104,105,106],
                   'name':['Vinay','Liyan','Farzana','Alia','Kanae','Hina']})
df
```

```python
df1 = pd.DataFrame({'id':[101,102,103,104,201,202],
                   'city':['pune','jakarta','tenuku','ajman','mumbai','hedrabad'],
                   'sal':[25008363000,909821,737393,2748739,94172164,76767]})
df1
```

```python
# merge
```

```python
pd.merge(df,df1,on='id')
```

```python
pd.merge(df,df1,on='id')[['name','city',]]
```

```python

pd.merge(df,df1,on='id')[['name','city','sal']]
```

```python
pd.merge(df,df1,on='id',indicator=True)
```

```python
# left join merge
```

```python
pd.merge(df,df1,on='id',how='left',indicator=True)
```

```python
# right merge/join
```

```python
pd.merge(df,df1,on='id',how='right',indicator=True)
```

```python
# outer join/merge
```

```python
pd.merge(df,df1,on='id',how='outer',indicator=True)
```
